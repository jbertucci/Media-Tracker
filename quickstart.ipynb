# Media Tracker — Quickstart

This notebook covers setup, starting the server, and data management utilities.

---

## 1. Start the server

Run the following in a terminal (not here). Keep it running while using the app:

```
cd "<project directory>"
py -3.9 api.py
```

---

## 2. Access the app

| Location | URL |
|---|---|
| This machine | http://localhost:5000 |
| Local network | http://\<local IP\>:5000 |
| Anywhere (Tailscale) | http://\<Tailscale IP\>:5000 |
| VM (always on) | http://\<VM IP\>:5000 |

When prompted, enter the API key from `flask_api_key.txt`.

---

## 3. Deploy updates to the VM

After committing and pushing changes:
```bash
# SSH into the VM
ssh -i "<path to SSH key>" ubuntu@<VM IP>

# Pull and restart
cd ~/Media-Tracker && git pull && sudo systemctl restart media-tracker
```

To push the local database to the VM:
```powershell
scp -i "<path to SSH key>" "<project directory>\tmdb_analytics.db" ubuntu@<VM IP>:~/Media-Tracker/
```
